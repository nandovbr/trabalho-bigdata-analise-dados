```python
for dados in dados_jogadores.itertuples():
    #dados
    nom = dados.jogador
    gen = dados.genero
    dat_nasc = dados.data_nascimento
    num_vit = dados.num_vitorias
    num_der = dados.num_derrotas
    tot_part = dados.total_partidas
    cod_pa = dados.pais
    cod_jog = dados.jogo

    # transforma formato da data para Y-m-d
    dat_nasc = datetime.strptime(dat_nasc, '%d/%m/%Y').date()
    #print(dados)
    try:
        #FOREIGN KEY (`cod_pais`) REFERENCES `data_players`.`pais` (`cod_pais`), FOREIGN KEY (`cod_jogo`) REFERENCES `data_players`.`jogo` (`cod_jogo`)
        sql = f"""INSERT INTO jogador (nome_jogador, genero, data_nascimento, num_vitorias, num_derrotas, total_partidas, cod_pais, cod_jogo) VALUES ('{nom}', '{gen}', '{dat_nasc}', '{num_vit}', '{num_der}', '{tot_part}', '{cod_pa}', '{cod_jog}')"""
        #print(sql)
        conn.execute(sql)
    except Exception as e:
        print(f'Erro ao inserir na tabela jogador: {e}')
        
```
