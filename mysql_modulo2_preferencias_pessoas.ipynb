```python
# criando a conexão com o banco de dados
# OBS: essas informações sempre ficam encapsuladas. Está sendo feito aqui para fim didático
user = 'root'
password = 'root'
host = '127.0.0.1' # local do host
port = '3307' # port
database = 'coletaIGTI' # schema criado no mysql workbanch

string_conexao = f'mysql://{user}:{password}@{host}:{port}/{database}'
string_conexao
```

```python
# engine de conexão com o servidor
engine = sqlalchemy.create_engine(string_conexao)
conn = engine.connect()
```

```python
# adicionando a tabela de pessoa no banco de dados workbench
for pessoa in df_pessoa.itertuples():
    pessoa
    codp = pessoa.cod_pessoa
    gen = pessoa.genero
    datanasc = pessoa.data_nascimento
    try:
        sql = f"""INSERT INTO pessoa (cod_pessoa, genero, data_nascimento) VALUES ('{codp}', '{gen}', '{datanasc}')"""
        #print(sql) testanto a saída do sql
        conn.execute(sql)
    except Exception as e:
        print(f'Erro ao inserir na tabela pessoa: {e}')
```

```python
# adicionando a tabela de animal_estimacao no banco de dados workbench
for animal in df_animal_estimacao.itertuples():
    #print(animal) testa a saída animal
    animal
    name_animal = animal.animal_estimacao
    # não é necessário capturar o valor de cod_animal_estimacao pois ele é auto incrementado no banco de dados
    try:
        sql = f"""INSERT INTO animal_estimacao (animal) VALUES ('{name_animal}')"""
        #print(sql) testando a saída do sql
        conn.execute(sql)
    except Exception as e:
        print(f'Erro ao inserir na tabela pessoa: {e}')
```

```python
# adicionando a tabela de clima no banco de dados workbench
for clima in df_clima.itertuples():
    #print(clima) testa a saída clima
    clima
    cli = clima.clima
    # não é necessário capturar o valor de cod_clima pois ele é auto incrementado no banco de dados
    try:
        sql = f"""INSERT INTO clima (clima) VALUES ('{cli}')"""
        #print(sql) testando a saída do sql
        conn.execute(sql)
    except Exception as e:
        print(f'Erro ao inserir na tabela pessoa: {e}')
```

```python
# adicionando a tabela de bebida no banco de dados workbench
for bebida in df_bebida_favorita.itertuples():
    #print(bebida) testa a saída bebida
    bebida
    beb = bebida.bebida_favorita
    # não é necessário capturar o valor de cod_clima pois ele é auto incrementado no banco de dados
    try:
        sql = f"""INSERT INTO bebida (bebida) VALUES ('{beb}')"""
        #print(sql) testando a saída do sql
        conn.execute(sql)
    except Exception as e:
        print(f'Erro ao inserir na tabela pessoa: {e}')
```
